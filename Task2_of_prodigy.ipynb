```python
# ✅ Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# ✅ Step 2: Upload train.csv
from google.colab import files
uploaded = files.upload()

df = pd.read_csv("train.csv")  # Replace if your file name is different
df.head()
```

```python
# ✅ Step 3: Check columns
df.columns
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Sex', palette='pastel')
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(data=df, x='Age', bins=10, kde=True, color='skyblue')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```
